```python
# from libraries
import pandas as pd
import dask.dataframe as dd
from io import StringIO
import boto3
import cStringIO
# from files
import params
import auth
import get_data_util
import get_sp500_util as sp500
```

```python
ticker_list = sp500.get_sp_500(url=sp500.url, headers=auth.user_agent)
quandl_parameters = [a for a in dir(params) if ("_parameters" in a)]
```

```python
alpha_vantage_data = get_data_util.get_alpha_vantage_data(ticker_list[:3])
```

```python
alpha_vantage_data
```

```python
quandl_data = get_data_util.get_quandl_data(quandl_parameters)
```

```python
stock_market_data = dd.merge(alpha_vantage_data, quandl_data)
```

```python
csv_buffer = StringIO()
stock_market_data.compute().to_csv(csv_buffer)
s3_resource = boto3.resource('s3')
s3_resource.Object(auth.bucket_name, '{}.csv'.format("tempfile")).put(Body=csv_buffer.getvalue())
```
